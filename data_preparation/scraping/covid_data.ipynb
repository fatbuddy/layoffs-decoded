```python
import pandas as pd
```

```python
import requests

url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
response = requests.get(url)

with open("owid-covid-data.csv", "wb") as f:
    f.write(response.content)
```

```python
covid_df = pd.read_csv('owid-covid-data.csv')
```

```python
covid_df.head()
```

```python
#Filter the data for only United States
US_covid_df = covid_df[covid_df['location']=='United States']
```

```python
US_covid_df.head()
```

```python
US_covid_df = US_covid_df.dropna(subset=["total_cases"])
```

```python
US_covid_df.columns
```

```python
US_covid_df_q = US_covid_df.copy()
US_covid_df_q["date"] = pd.to_datetime(US_covid_df_q["date"])
US_covid_df_q = US_covid_df_q.groupby(pd.Grouper(key="date", freq='Q'))\
                            [["new_cases","new_cases_smoothed","new_cases_per_million",\
                              "new_deaths","new_deaths_smoothed","new_deaths_per_million"]].mean()
US_covid_df_q = US_covid_df_q.reset_index()
display(US_covid_df_q)
```

```python
US_covid_df_q.to_csv("quarterly_covid_data.csv")
```
